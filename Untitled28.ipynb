```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
nifty = pd.read_csv('NIFTY 100_minute.csv')
```

```python
nifty['date'] = pd.to_datetime(nifty['date'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
nifty.set_index('date', inplace=True)
```

```python
plt.figure(figsize=(12,6))
plt.plot(nifty.index, nifty['close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('NIFTY 100 Close Price')
# Convert year values to integers using astype(int) to avoid the TypeError
for year in range(int(min(nifty.index.year)), int(max(nifty.index.year))+1):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--')
plt.show()
```

```python
daily_data = nifty.resample('D').mean()
daily_data['close_diff'] = daily_data['close'].diff().dropna()
adf_test = adfuller(daily_data['close_diff'].dropna())
print("ADF Statistic:", adf_test[0])
print("p-value:", adf_test[1])
print("Critical Values:", adf_test[4])
```

```python
plt.figure(figsize=(12,6))
plt.plot(daily_data.index, daily_data['close_diff'])
plt.xlabel('Date')
plt.ylabel('Daily Close Price Difference')
plt.title('NIFTY 100 Daily Close Price Difference')
for year in range(min(daily_data.index.year), max(daily_data.index.year)+1):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--')
```

```python
plt.figure(figsize=(12, 6))

# ACF Plot
plt.subplot(2, 1, 1)
plot_acf(daily_data['close_diff'].dropna(), lags=40, ax=plt.gca())
plt.title("ACF (Autocorrelation Function)")

# PACF Plot
plt.subplot(2, 1, 2)
plot_pacf(daily_data['close_diff'].dropna(), lags=40, ax=plt.gca(), method="ywm")
plt.title("PACF (Partial Autocorrelation Function)")

plt.tight_layout()
plt.show()
```

```python
# Define the size of the training set (80% of the data)
train_size = int(len(daily_data) * 0.8)

# Split the data into train and test sets
train = daily_data['close'][:train_size]
test = daily_data['close'][train_size:]
```

```python

model = ARIMA(train, order=(1, 1, 1))  # Adjust order=(p, d, q) based on ACF/PACF plots
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())
```

```python
# One-step predictions for the test set
predictions = model_fit.forecast(steps=len(test))

# Rolling Forecast
rolling_predictions = []
history = list(train)

for t in test:
    # Fit the model on the history (past data)
    model = ARIMA(history, order=(1, 1, 1))
    model_fit = model.fit()

    # Forecast the next value
    yhat = model_fit.forecast(steps=1)[0]
    rolling_predictions.append(yhat)

    # Append the true value to history
    history.append(t)

# Convert rolling predictions to a Pandas series
rolling_predictions = pd.Series(rolling_predictions, index=test.index)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train, label='Training Data', color='blue')
plt.plot(test, label='Actual Test Data', color='green')
plt.plot(rolling_predictions, label='Rolling Forecast', color='red')
plt.title('Rolling Forecast vs Actuals')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.plot(test, label='Actual Test Data', color='green')
plt.plot(rolling_predictions, label='Rolling Forecast', color='red')
```

```python
#Model Accuracy
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Remove NaN values from both test and rolling_predictions
test_clean = test.dropna()
rolling_predictions_clean = rolling_predictions.dropna()

# Align the indices of the two series
common_index = test_clean.index.intersection(rolling_predictions_clean.index)
test_clean = test_clean.loc[common_index]
rolling_predictions_clean = rolling_predictions_clean.loc[common_index]

#Mean Absolute Error
mae = mean_absolute_error(test_clean, rolling_predictions_clean)
print("Mean Absolute Error:", mae)

#Mean Squared Error
mse = mean_squared_error(test_clean, rolling_predictions_clean)
print("Mean Squared Error:", mse)

#Root Mean Squared Error
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)

#R-squared
r2 = r2_score(test_clean, rolling_predictions_clean)
print("R-squared:", r2)  # Corrected the label to R-squared

#Means Absolute Percentage Error
mape = np.mean(np.abs((test_clean - rolling_predictions_clean) / test_clean)) * 100
print("Mean Absolute Percentage Error:", mape)
```

```python
#Future Predictions
future_steps = 30
forecast = model_fit.forecast(steps=future_steps).round(2)
print(forecast)
plt.figure(figsize=(12, 6))
plt.plot(daily_data.index, daily_data['close'], label='Historical Data')
plt.plot(pd.date_range(start=daily_data.index[-1], periods=future_steps+1, freq='D')[1:], forecast, label='Future Predictions')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Historical Data and Future Predictions')
plt.legend()
plt.grid()
plt.show()

```
